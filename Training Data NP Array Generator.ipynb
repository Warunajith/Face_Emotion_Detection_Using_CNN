```python
import cv2,os


data_path=r'C:\Users\warunajith\Documents\CNN Models\data\train'
categories=os.listdir(data_path)
labels=[i for i in range(len(categories))]

label_dict={} #empty dictionary

for i in range(len(categories)):
    label_dict[categories[i]]=labels[i]

print(label_dict)
print(categories)
print(labels)
```

```python
img_size=48

train_data=[]
train_target=[]

for category in categories:
    folder_path=os.path.join(data_path,category)
    img_names=os.listdir(folder_path)
    
    for img_name in img_names:
        img_path=os.path.join(folder_path,img_name)
        img=cv2.imread(img_path)
        
        try:
            gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
            train_data.append(gray)
            train_target.append(label_dict[category])
            
        except Exception as e:
            print(e)
```

```python
import numpy as np

train_data=np.array(train_data)
train_target=np.array(train_target)
```

```python
print(train_data.shape,train_target.shape)
```

```python
train_data_new=train_data/255.0
print(train_data_new.shape)
train_data_new=np.reshape(train_data_new,(train_data_new.shape[0],img_size,img_size,1))

from keras.utils import np_utils

train_target_new=np_utils.to_categorical(train_target)
print(train_target_new.shape)
```

```python
np.save('train_data',train_data_new)
np.save('train_target',train_target_new)
```
